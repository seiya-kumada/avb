```python
import numpy as np
import os
import matplotlib.pyplot as plt
import math
from constants import *  # noqa
%matplotlib inline
```

# display leaning curve

```python
DIR_PATH = './result'
epoch_phi_losses = np.load(os.path.join(DIR_PATH, 'epoch_phi_losses.npy'))
epoch_psi_losses = np.load(os.path.join(DIR_PATH, 'epoch_psi_losses.npy'))
epoch_kls = np.load(os.path.join(DIR_PATH, 'epoch_kls.npy'))
```

```python
plt.plot(epoch_phi_losses, label='phi')
plt.plot(epoch_psi_losses, label='psi')
plt.plot(epoch_kls, label='kl')
plt.hlines([0, math.log(4)], 0, 400)
# plt.ylim(-0.5, 0.5)
plt.legend(loc='best')
plt.show()
```
